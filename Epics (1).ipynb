```python
import os
import xml.etree.ElementTree as ET
from tqdm import tqdm
from PIL import Image
```

```python
base_path = "/content/NTLNP"
output_path = "/content/yolo_dataset"
splits = ["day", "night"]
classes = []
```

```python
def convert_voc_to_yolo(xml_file, img_w, img_h):
    tree = ET.parse(xml_file)
    root = tree.getroot()

    yolo_data = []

    for obj in root.findall("object"):
        cls_name = obj.find("name").text
        if cls_name not in classes:
            classes.append(cls_name)
        cls_id = classes.index(cls_name)

        bbox = obj.find("bndbox")
        xmin = int(bbox.find("xmin").text)
        xmax = int(bbox.find("xmax").text)
        ymin = int(bbox.find("ymin").text)
        ymax = int(bbox.find("ymax").text)

        # Convert to YOLO format
        x_center = ((xmin + xmax) / 2) / img_w
        y_center = ((ymin + ymax) / 2) / img_h
        width = (xmax - xmin) / img_w
        height = (ymax - ymin) / img_h

        yolo_data.append(f"{cls_id} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}")

    return yolo_data

for split in splits:
    ann_dir = os.path.join(base_path, split, "Annotations")
    img_dir = os.path.join(base_path, split, "JPEGImages")

    out_img_dir = os.path.join(output_path, "images", split)
    out_lbl_dir = os.path.join(output_path, "labels", split)
    os.makedirs(out_img_dir, exist_ok=True)
    os.makedirs(out_lbl_dir, exist_ok=True)

    for filename in tqdm(os.listdir(img_dir)):
        if not filename.endswith(".jpg"):
            continue

        img_path = os.path.join(img_dir, filename)
        xml_path = os.path.join(ann_dir, filename.replace(".jpg", ".xml"))
        if not os.path.exists(xml_path):
            continue


        with Image.open(img_path) as img:
            w, h = img.size


        yolo_lines = convert_voc_to_yolo(xml_path, w, h)


        label_path = os.path.join(out_lbl_dir, filename.replace(".jpg", ".txt"))
        with open(label_path, "w") as f:
            f.write("\n".join(yolo_lines))


        os.system(f"cp '{img_path}' '{out_img_dir}'")


with open(os.path.join(output_path, "classes.txt"), "w") as f:
    f.write("\n".join(classes))

print(f"Conversion done! Classes: {classes}")
```

```python
import os
import shutil
import random
from tqdm import tqdm

# Paths
base_path = "/content/yolo_dataset"
image_day = os.path.join(base_path, "images/day")
image_night = os.path.join(base_path, "images/night")
label_day = os.path.join(base_path, "labels/day")
label_night = os.path.join(base_path, "labels/night")

# Output folders
out_images = os.path.join(base_path, "images")
out_labels = os.path.join(base_path, "labels")
for split in ["train", "val"]:
    os.makedirs(os.path.join(out_images, split), exist_ok=True)
    os.makedirs(os.path.join(out_labels, split), exist_ok=True)

# Collect all image-label pairs
all_images = []
for subfolder in [image_day, image_night]:
    for file in os.listdir(subfolder):
        if file.endswith(".jpg"):
            img_path = os.path.join(subfolder, file)
            label_path = os.path.join(base_path, "labels", os.path.basename(subfolder), file.replace(".jpg", ".txt"))
            if os.path.exists(label_path):
                all_images.append((img_path, label_path))

# Shuffle and split
random.shuffle(all_images)
split_idx = int(0.8 * len(all_images))
train_data = all_images[:split_idx]
val_data = all_images[split_idx:]

# Function to copy
def copy_to_split(pairs, split):
    for img, label in tqdm(pairs, desc=f"Copying to {split}"):
        shutil.copy(img, os.path.join(out_images, split, os.path.basename(img)))
        shutil.copy(label, os.path.join(out_labels, split, os.path.basename(label)))

copy_to_split(train_data, "train")
copy_to_split(val_data, "val")
```

```python
class_names = [
    "Amur_tiger", "Amur_leopard", "wild_boar", "roe_deer", "sika_deer",
    "Asian_black_bear", "red_fox", "Asian_badger", "raccoon_dog", "musk_deer",
    "Siberian_weasel", "sable", "yellow_throated_marten", "leopard_cat", "Manchurian_hare",
    "cow", "dog"
]

yaml_content = f"""
train: /content/yolo_dataset/images/train
val: /content/yolo_dataset/images/val

nc: {len(class_names)}
names: {class_names}
"""

with open("/content/yolo_data.yaml", "w") as f:
    f.write(yaml_content)

print("data.yaml created!")
```

```python
import os
import random
import shutil


SOURCE_IMAGES = '/content/yolo_dataset/images'
SOURCE_LABELS = '/content/yolo_dataset/labels'
OUTPUT_DIR = '/content/mini_dataset'
SAMPLE_RATIO = 0.4
TRAIN_RATIO = 0.8


image_files = []
for root, _, files in os.walk(SOURCE_IMAGES):
    for file in files:
        if file.endswith(('.jpg', '.jpeg', '.png')):
            image_files.append(os.path.join(root, file))


sampled_images = random.sample(image_files, int(len(image_files) * SAMPLE_RATIO))
random.shuffle(sampled_images)


split_index = int(len(sampled_images) * TRAIN_RATIO)
train_images = sampled_images[:split_index]
val_images = sampled_images[split_index:]


for subset in ['train', 'val']:
    for subfolder in ['images', 'labels']:
        os.makedirs(os.path.join(OUTPUT_DIR, subfolder, subset), exist_ok=True)


def copy_files(image_list, subset):
    for img_path in image_list:
        img_name = os.path.basename(img_path)
        dest_img = os.path.join(OUTPUT_DIR, 'images', subset, img_name)
        shutil.copy2(img_path, dest_img)


        rel_path = os.path.relpath(img_path, SOURCE_IMAGES)
        label_path = os.path.join(SOURCE_LABELS, os.path.splitext(rel_path)[0] + '.txt')
        dest_label = os.path.join(OUTPUT_DIR, 'labels', subset, os.path.splitext(img_name)[0] + '.txt')

        if os.path.exists(label_path):
            os.makedirs(os.path.dirname(dest_label), exist_ok=True)
            shutil.copy2(label_path, dest_label)


copy_files(train_images, 'train')
copy_files(val_images, 'val')

print(f"mini_dataset created with {len(train_images)} training and {len(val_images)} validation images.")
```

```python
class_names = [
    "Amur_tiger", "Amur_leopard", "wild_boar", "roe_deer", "sika_deer",
    "Asian_black_bear", "red_fox", "Asian_badger", "raccoon_dog", "musk_deer",
    "Siberian_weasel", "sable", "yellow_throated_marten", "leopard_cat", "Manchurian_hare",
    "cow", "dog"
]

yaml_content = f"""
train: /content/mini_dataset/images/train
val: /content/mini_dataset/images/val

nc: {len(class_names)}
names: {class_names}
"""

with open("/content/mini_data.yaml", "w") as f:
    f.write(yaml_content)

print("data.yaml created!")
```
